```python
input = [[1,1],[1,-1],[-1,1],[-1,-1]]
output = [-1,1,-1,-1]
alpha = b = 0.2
w1 = w2 = 0.2

def Addaline(input,output,w1,w2,b):
  for i in range(5):
    print("Epoch",i+1)
    print("x1 \t x2 \t y \t w1 \t w1 \t bias \t Error")
    for (x1,x2),y in zip(input, output):
      te=0
      yin = b+x1*w1+x2*w2
      e = (y-yin)
      w1=w1+alpha*e*x1
      w2=w2+alpha*e
      b=b+e
      E=e*e
      print(x1,x2,y,w1,w2,b,E)
      print(f'{round(x1,3)} \t  {round(x2,3)} \t {round(y,3)} , \t {round(w1,3)} \t {round(w2,3)} \t {round(b,3)} \t {round(E,3)}')
      te = E +te
      mean_error = te/2
      print(f'Total Mean Error After {i} Epochs:{mean_error}')
Addaline(input,output,w1,w2,b)
```
