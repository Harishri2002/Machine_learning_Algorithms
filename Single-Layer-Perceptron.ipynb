```python
input = [[1,1],[1,-1],[-1,1],[-1,-1]]
output = [1,-1,-1,-1]

w1=w2=b=0
alpha = 1
theta = 0

def singleLayer(input,output,w1,w2,b,alpha,theta):
  for epoch in range(5):
    print("Epoch:",epoch)
    for (x1,x2),y in zip(input,output):
      net_input = b+x1*w1+x2*w2
      if(net_input > theta):
        a=1
      elif(net_input < theta):
        a=-1
      else:
        a=0
      if a!=y:
        w1 += alpha * x1 * y
        w2 += alpha * x2 * y
        b += alpha * y
      print(f"input:((x1,x2)),target:(y),output:{a},w1:{w1},w2{w2},b:{b}")
singleLayer(input,output,w1,w2,b,alpha,theta)
```
