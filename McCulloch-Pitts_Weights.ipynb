```python
import numpy as np
def mcculloch(inputs,weights,threshold):
  y=np.dot(inputs,weights)
  if(y >= threshold):
    return 1
  else:
    return 0
input_values=[(0,0),(0,1),(1,0),(1,1)]
output_values=[0,0,1,0]
weights_list = [[1,1],[1,-1],[-1,1],[-1,-1]]
threshold = 1

print("Evaluating AND-NOT gate with different weight configurations:\n")
for weights in weights_list:
  print(f'Weights:{weights}')
  matches = True
  for inputs,expected in zip(input_values,output_values):
    result = mcculloch(inputs,weights,threshold)
    print(f'inputs:{inputs},Expected:{expected},Result:{result}')
    if(result!=expected):
      matches = False
    print(f"Matches AND-NOT Gate: {'Yes' if matches else 'No'}\n")
```
