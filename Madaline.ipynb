```python
def Madaline(input,output,alpha,v1,v2,v3,w11,w12,b1,w21,w22,b2):
  def actfunc(x):
    if x>=0:
      return 1
    return -1
  for i in range(5):
    print("EPOCH",i+1)
    print("x1 \t x2 \t zin1 \t zin2 \t z1 \t z2 \t y \t w11 \t w12 \t b1 \t w21 \t w22 \t b2")
    for (x1,x2),t in zip(input,output):
      zin1 = b1 + x1 * w11 + x2 * w21
      zin2 = b2 + x2 * w12 + x2 * w22
      zin = [zin1,zin2]
      z1 = actfunc(zin1)
      z2 = actfunc(zin2)

      yin = b3 + z1 * v1 + z2 * v2
      y = actfunc(yin)

      if y!=t:
        if t==1:
          if abs(zin1)<abs(zin2):
            w11 += alpha * (t-zin1)* x1
            w21 += alpha * (t-zin1)* x2
            b1 += alpha * (t-zin1)
          else:
            w12 += alpha * (-t-zin2)* x1
            w22 += alpha * (-t-zin2)* x2
            b2 += alpha * (t-zin2)
        else:
          if abs(zin1)<abs(zin2):
            w11 += alpha * (-t-zin1)* x1
            w21 += alpha * (-t-zin1)* x2
            b1 += alpha * (-t-zin1)
          else:
            w12 += alpha * (-t-zin2)* x1
            w22 += alpha * (-t-zin2)* x2
            b2 += alpha * (-t-zin2)
      print(f"{round(x1,3)} \t {round(x2,3)} \t {round(zin1,3)} \t {round(zin2,3)} \t {round(z1,3)} \t {round(z2,3)} \t {round(y,3)} \t {round(w11,3)} \t {round(w12,3)} \t {round(b1,3)} \t {round(w21,3)} \t {round(w22,3)} \t {round(b2,3)}")


#Madaline Network
input = [[1,1],[1,-1],[-1,1],[-1,-1]]
output = [-1,1,1,-1]
alpha = 0.5
v1=v2=v3=b3=0.5
w11 = 0.05
w12 = 0.2
b1 = 0.3

w21 = 0.1
w22 = 0.2
b2 = 0.15

Madaline(input,output,alpha,v1,v2,v3,w11,w12,b1,w21,w22,b2)
```
